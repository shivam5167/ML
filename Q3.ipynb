```python
for i in df.columns:
    print(f"{i[0:4]:.^5}",end='|')
print('\n')
for i in range (df.shape[0]):
    for j in range(df.shape[1]):
            if(type(df.iloc[i,j])==str):
                print(f"{df.iloc[i,j][0:4]:.^5}",end='|')
            elif(type(df.iloc[i,j])==np.int64):
                print(f"{df.iloc[i,j]:.<5}",end='|')
            else:
                print(f"{df.iloc[i,j]:.>5}",end='|')
    print("\n")
            
   
```
