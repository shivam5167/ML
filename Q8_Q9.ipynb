```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=datasets.load_iris()
x=pd.DataFrame(iris.data)
x.columns=iris.feature_names
x["species"]=iris.target
x.head()
```

```python
#BOXPLOT
plt.boxplot(iris.data)
plt.show()
```

```python
#HISTOGRAM
sl = x['sepal length (cm)']
plt.hist(x,np.arange(sl.min(), sl.max(), 4))
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
sns.pairplot(x)
```

```python
plt.hist(x["sepal length (cm)"],bins=6)
```

```python
plt.pie(x["species"].value_counts(),labels=iris.target_names)
plt.show()
```

```python
plt.bar(iris.target_names,x["species"].value_counts(),color=["red","green","blue"])
```

```python
#sin cos
import math
d=np.arange(-360,360)
r=[math.radians(i) for i in d]
fig,axes=plt.subplots(2,1)
axes[0].plot(d,np.sin(r))
axes[1].plot(d,np.cos(r))
```
